```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/brianna-r-_-54wJzpH6Q-unsplash.jpg')

img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
m,n=img.shape
gaussian_noise=np.zeros((m,n),dtype=np.uint8)
gaussian_noise

cv2.randn(gaussian_noise,122,23)
guassian_noise=(gaussian_noise*0.5).astype("uint8")


gaussian_img=cv2.add(img,gaussian_noise)

plt.subplot(1,2,1)
plt.imshow(guassian_noise,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(gaussian_img,cmap='gray')

plt.show()
```
