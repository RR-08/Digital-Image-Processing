```python
# Gaussian Blur

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/nicholas-selman-fRu0x7L6-H0-unsplash.jpg')

blur = cv2.GaussianBlur(img, (5,5), 2)
blur1 = cv2.GaussianBlur(img, (11,11), 2)
blur2 = cv2.GaussianBlur(img, (15,15), 2)

plt.subplot(221), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB),cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('Kernel size:5')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(blur1, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('Kernel size:11')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(blur2, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('Kernel size:15')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#changing standard deviation
blur = cv2.GaussianBlur(img, (15,15), 0.2)
blur1 = cv2.GaussianBlur(img, (15,15), 2)
blur2 = cv2.GaussianBlur(img, (15,15), 1)

plt.subplot(221), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB),cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('sd=0.2')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(blur1, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('sd=2')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(blur2, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('sd=1')
plt.xticks([]), plt.yticks([])
```

```python
#Mean Blur
img = cv2.imread('/content/nicholas-selman-fRu0x7L6-H0-unsplash.jpg')
blur = cv2.blur(img, (15,15))

plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB),cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB),cmap="gray"), plt.title('Averaging')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#median filter

blur = cv2.medianBlur(img, 5)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur), plt.title('Averaging')
plt.xticks([]), plt.yticks([])

plt.show()
```
