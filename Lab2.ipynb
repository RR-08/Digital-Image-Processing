```python
import numpy
import cv2
import sklearn
from skimage import color
from skimage import io,transform
from matplotlib import pyplot as plt
img1 = io.imread("/content/img1.jpg")
img2= io.imread("/content/img2.jpg")
fig=plt.figure()
ax1=fig.add_subplot(1,2,1)
ax1.imshow(img1)
ax2=fig.add_subplot(1,2,2)
ax2.imshow(img2)
```

```python

resizedimg2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
# plt.imshow(resizedimg2)
addimg=cv2.add(img1,resizedimg2)
plt.imshow(addimg)
```

```python
resizedimg2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
# addimg2=cv2.addWeighted(img1,0.4,resizedimg2,0.6,0.1)
addimg2=img1+resizedimg2
plt.imshow(addimg2)
```

```python
resizedimg2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
addimg2=cv2.addWeighted(img1,0.4,resizedimg2,0.6,0.1)
plt.imshow(addimg2)
```

```python
#addition
resizedimg2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
addimg2=img1+resizedimg2
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.imshow(addimg2)
ax2=fig.add_subplot(2,2,2)
ax2.imshow(cv2.add(img1,resizedimg2))
ax1=fig.add_subplot(2,2,3)
ax1.imshow(img1+1)
ax1=fig.add_subplot(2,2,4)
ax1.imshow(img2+0.1)
```

```python
#subtraction
resizedimg2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
subimg2=img1-resizedimg2
fig=plt.figure()
#numpy
ax1=fig.add_subplot(1,2,1)
ax1.imshow(subimg2)
ax2=fig.add_subplot(1,2,2)
ax2.imshow(cv2.subtract(img1,resizedimg2))
```

```python
#multiplication
mulimg2=img1*resizedimg2
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.imshow(mulimg2)
ax2=fig.add_subplot(2,2,2)
ax2.imshow(cv2.multiply(img1,resizedimg2))
ax3=fig.add_subplot(2,2,3)
ax3.imshow(img1*5)
ax4=fig.add_subplot(2,2,4)
ax4.imshow(img2*28)
```

```python
#division
divimg2=img1*resizedimg2
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.imshow(divimg2)
ax2=fig.add_subplot(2,2,2)
ax2.imshow(cv2.divide(resizedimg2,img1))
ax3=fig.add_subplot(2,2,3)
ax3.imshow(img1/6)
ax4=fig.add_subplot(2,2,4)
ax4.imshow(img2/8)
```
