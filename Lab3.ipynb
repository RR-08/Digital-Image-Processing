# Histogram


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread("/content/brianna-r-_-54wJzpH6Q-unsplash.jpg" , cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found.")
else:
    modified_image = cv2.add(image, 120)

    original_hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    modified_hist = cv2.calcHist([modified_image], [0], None, [256], [0, 256])

    plt.subplot(121)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
    plt.title('Modified Image')


    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.plot(original_hist)
    plt.title('Original Image Histogram')

    plt.subplot(122)
    plt.plot(modified_hist)
    plt.title('Modified Image Histogram')



    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.plot(original_hist)
    plt.title('Original Image Histogram')

    plt.subplot(122)
    plt.plot(modified_hist)
    plt.title('Modified Image Histogram')

    plt.show()
```
